```python
import pandas as pd
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary
from gensim import models
import pyLDAvis.gensim
import pyLDAvis.sklearn
import os
```

```python
os.chdir("/Users/lucasgomes/Documents/projetos/Essex/MicrosoftEssexLDA")
df = pd.read_csv("../MicrosoftEssexScrapers/results/CNN/articles.csv")
```

```python
df.Text = df.Text.apply(lambda x: x.split())
common_texts = df.Text.tolist()
```

```python
# Create a corpus from a list of texts
common_dictionary = Dictionary(common_texts)
common_corpus = [common_dictionary.doc2bow(text) for text in common_texts]
# Train the model on the corpus.
lda = models.LdaModel(common_corpus, num_topics=10)
```

```python
p = pyLDAvis.gensim.prepare(lda, common_corpus, common_dictionary, n_jobs=-1, sort_topics=False)

pyLDAvis.save_html(p, "results/FONTE".replace("FONTE","LDA.html"))
```
